```python
my_file = open("file.txt", "r")
content_list = my_file.readlines()
#print(content_list)
print(content_list[1])
names=content_list
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('french'))
```

```python
import spacy
from spacy import displacy
nlp = spacy.load('fr_core_news_sm')
```

```python
def clean_text(text):
    text = text.lower() # mettre les mots en minuscule
    text = " ".join(text.split())
    return text
```

```python
dict = []
for elem in names:
    elem = elem.replace('-', ' ')
    elem = nltk.word_tokenize(elem, language='french')
        
    dict.append(elem)
    
print(dict)
```

```python
def lemmatize_spacy(sentence):
    nlp = spacy.load("fr_core_news_sm")
    doc=nlp(sentence)
    output=[]
    for elem in doc:
        output.append(elem.lemma_)
    return output

print(lemmatize_spacy('la souris mange le chat'))
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer(language='french')
```

```python
def lemmatize_ntlk(sentence):
    stemmer = SnowballStemmer(language='french')
    output=nltk.word_tokenize(sentence, language='french')
    vector=[]
    for elem in output:
        lemma = elem.replace(elem,stemmer.stem(elem))
        vector.append(lemma)
    return vector
        
print(lemmatize_ntlk('la souris mange le chat'))
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer(language='french')
liste_racines=[]
for elem in dict:
    racines=[]
    liste_racines.append(racines)
    for word in elem:
        toadd=word.replace(word,stemmer.stem(word))
        racines.append(toadd)

print(liste_racines)
```

```python
def match_outcome_with_list(ingredient):
    sentence=clean_text(ingredient)
    output=lemmatize_ntlk(sentence)
    print(output)
    count=0
    closest_match=[]
    for elem in liste_racines:
        increment=0
        for token in elem:
            if token in output:
                increment+=1
                if count<=increment:
                    count=increment
                    closest_match=elem
    if count == 0:
        closest_match=['ingredient non trouve']
    return dict[liste_racines.index(closest_match)]
        
```
